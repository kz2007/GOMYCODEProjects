```python
def getKey(item):
    return item[1]

List = [(2, 5), (1, 2), (4, 4), (2, 3), (2, 1)]
print(sorted(List, key=getKey))
```

```python
d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}
d3 = dict(d1)
d3.update(d2)



for x, y in d1.items():
    for a, b in d2.items():
        if a == x:
            d3[a] = y + b
    
            
print(d3)
    
```

```python
n = input("Enter an integer: ")
n = int(n)
d = {}
for i in range(1, n+1):
    d[i] = i**2

print(d)
    
```
